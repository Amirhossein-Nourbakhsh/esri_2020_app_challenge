```python
import requests
import pandas as pd
```

```python
hospitals = []
df_hospitals = pd.read_csv('Hospital Spread sheet - Copy of Sheet1.csv')
```

```python
ratings = []
```

```python
for hospital_name in df_hospitals['Facility Name']:
    api_url = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input=' \
            + hospital_name + \
            '&inputtype=textquery&fields=formatted_address,name,rating&locationbias=circle:2000@43.72,-79.49&key=AIzaSyDX7jEBDK9qjnWaJbHbJKtX9R25Uiy6ny8'
    response = requests.get(url=api_url)
    data = response.json()
    rating = data['candidates'][0]['rating']
    ratings.append(rating)
```

```python
len(ratings)
```

```python
df_hospitals['yelp ratings'] = ratings
```

```python
df_hospitals = df_hospitals.drop(columns = ['Yelp Rating', 'Insurance Supportive', 'Available Emergency Services', 
                                            'User Feedback', 'CommentIDs', 'Unnamed: 13', 'OFFICE_PHONE', 
                                            'EMAIL', 'WEBSITE'])
```

```python
df_hospitals = df_hospitals.drop(columns = ['Geometry', 'Doctors Number'])
```
